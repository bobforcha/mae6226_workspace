# Lesson 2: Source and Sink in  a Freestream

Using superposition, we will add the solutions of a source, sink and uniform freestream

## Code:

```python
# imports
import math
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# create meshgrid
N = 200
x_start, x_end = -4.0, 4.0
y_start, y_end = -2.0, 2.0
x = np.linspace(x_start,x_end,N)
y = np.linspace(y_start,y_end,N)
X, Y = np.meshgrid(x,y)
```

## Source in a uniform flow

The streamlines of a freestream with speed $U_\infty$ and angle of attack $\alpha$ are given by:

$$\psi_\text{freestream}\left(x\text{,} y\right) = U_\infty\left(y\cos{\alpha} - x\sin\alpha\right)$$

```python
u_inf = 1.0        # freestream speed

# calculate the freestream velocity field
u_freestream = u_inf * np.ones((N,N), dtype=float)
v_freestream = np.zeros((N,N), dtype=float)

# compute the free stream function
psi_freestream = u_inf * Y
```

$$\psi_\text{source} \left(x\text{,} y\right)= \frac{\sigma}{2\pi}\arctan \left(\frac{y - y_\text{source}}{x - x_\text{source}}\right)$$

$$u_\text{source} \left(x \text{,} y\right) = \frac{\sigma}{2\pi} \frac{x - x_\text{source}}{\left(x - x_\text{source}\right)^2 + \left(y - y_\text{source}\right)^2}$$

$$v_\text{source} \left(x \text{,} y\right) = \frac{\sigma}{2\pi} \frac{y - y_\text{source}}{\left(x - x_\text{source}\right)^2 + \left(y - y_\text{source}\right)^2}$$

### Velocity Function

```python
def get_velocity(strength, xs, ys, X, Y):
    """
    PUNCHED IN THE FACE!!!
    
    you have been punched in the face.
    """
    
    u = strength/(2*math.pi)*(X - xs)/((X - xs)**2 + (Y - ys)**2)
    v = strength/(2*math.pi)*(Y - ys)/((X - xs)**2 + (Y - ys)**2)
    return u, v
```

### Stream Function

```python
def get_stream_function(strength, xs, ys, X, Y):
    """
    NOTHING TO SEE HERE!!!

    please go about your business
    """

    psi = strength/(2*math.pi)*np.arctan2((Y - ys), (X - xs))
    return psi
```

```python
strength_source = 5.0
x_source, y_source = -1.0, 0.0
u_source, v_source = get_velocity(strength_source, x_source, y_source, X, Y)
psi_source = get_stream_function(strength_source, x_source,y_source, X, Y)
```

### Superposition

$$\psi = \psi_\text{freestream} + \psi_\text{source}$$

$$u = u_\text{freestream} + u_\text{source}\text{,} \quad v = v_\text{freestream} + v_\text{source}$$

Stagnation points:

$$u = 0, \quad v = 0$$

$$x_\text{stag} = x_\text{source} - \frac{\sigma}{2\pi U_\infty}$$

$$y_\text{stag} = y_\text{source}$$

```python
u = u_freestream + u_source
v = v_freestream + v_source
psi = psi_freestream + psi_source

# plot the world's end
size = 10
plt.figure(figsize=(size,(y_end - y_start)/(x_end - x_start)*size))
plt.grid(True)
plt.xlabel('x', fontsize=16)
plt.ylabel('y', fontsize=16)
plt.xlim(x_start, x_end)
plt.ylim(y_start, y_end)
plt.streamplot(X, Y, u, v, density=2, linewidth=1, arrowsize=1, arrowstyle='->')
plt.scatter(x_source, y_source, color='#CD2305', marker='o', s=80)

# calculate the stagnation point
x_stag = x_source - strength_source/(2*math.pi*u_inf)
y_stag = y_source

# plot stagnation point
plt.scatter(x_stag, y_stag, color='g', marker='o', s=80)

# plot dividing streamline
plt.contour(X, Y, psi, levels=[-strength_source/2, +strength_source/2], colors='#CD2305', linewidths=2, linestyles='solid');
```

This is called a *Rankine half-body*

Conservation of mass dictates $U_\infty D = \sigma \quad \therefore \quad D = \frac{\sigma}{U_\infty}$ 

## Source-sink pair in a uniform flow

We now add a sink to the solution

```python
strength_sink = -5.0
x_sink, y_sink = 1.0, 0.0

u_sink, v_sink = get_velocity(strength_sink, x_sink, y_sink, X, Y)
psi_sink = get_stream_function(strength_sink, x_sink, y_sink, X, Y)

u = u_freestream + u_source + u_sink
v = v_freestream + v_source + v_sink
psi = psi_freestream + psi_source + psi_sink

# D = np.sqrt(strength_source*(strength_source/(math.pi*u_inf) + 4)/(math.pi*u_inf))

# plots
size = 10
plt.figure(figsize=(size,(y_end - y_start)/(x_end - x_start)*size))
plt.grid(True)
plt.xlabel('x', fontsize=16)
plt.ylabel('y', fontsize=16)
plt.xlim(x_start, x_end)
plt.ylim(y_start, y_end)
plt.streamplot(X, Y, u, v, density=2, linewidth=1, arrowsize=1, arrowstyle='->')
plt.scatter([x_source, x_sink], [y_source, y_sink], color='#CD2305', marker='o', s=80)
# plt.scatter(0,D/2,color='g', marker='o',s=80)
plt.contour(X, Y, psi, levels=[0.], colors='#CD2305', linewidths=2, linestyles='solid');
```

This is called a *Rankine Oval*

##### Challenge Question:

What is the length and width of the Rankine oval?

##### Answer:

The length of the Rankine oval is the distance between the source and the sink, plus the distance from the source to its stagnation point and the distance from the sink to its stagnation point.  If $\sigma_\text{source} = \sigma_\text{sink} = \sigma$:

$$l = \left|x_\text{source} - x_\text{sink}\right| + \left|\frac{\sigma}{\pi U_\infty}\right|$$

## Bernoulli's Equation and the Pressure Coefficient

Incompressible Bernoulli's:

$$p_\infty + \frac{1}{2}\rho V_{\infty}^2 = p + \frac{1}{2}\rho V^2$$

$$C_p = \frac{p - p_\infty}{\frac{1}{2}\rho V_{\infty}^2} = 1 - \left(\frac{V}{V_\infty}\right)^2$$

$C_p = 1$ at a stagnation point

## Code:

```python
# compute the pressure coefficient field
c_p = 1.0 - (u**2 + v**2)/u_inf**2

# plot the pressure coefficient field
plt.figure(figsize=(1.1*size,(y_end-y_start)/(x_end-x_start)*size))
plt.xlabel('x',fontsize=16)
plt.ylabel('y',fontsize=16)
plt.xlim(x_start, x_end)
plt.ylim(y_start, y_end)
contf = plt.contourf(X,Y,c_p,levels=np.linspace(-2.0,1.0,100),extend='both')
cbar = plt.colorbar(contf)
cbar.set_label('$C_p$',fontsize=16)
cbar.set_ticks([-2.0, -1.0, 0.0, 1.0])
plt.scatter([x_source,x_sink],[y_source,y_sink],color='#CD2305',s=80,marker='o')
plt.contour(X,Y,psi,levels=[0.],colors='#CD2305',linewidths=2,linestyles='solid');
```
